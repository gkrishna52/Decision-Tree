### <font color ='black'> Extracting Independent Variables </font>

### <font color ='black'> Splitting the dataset into training and test set </font>

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,y_train,x_test,x_train = train_test_split(X,Y, test_size = 0.25, random_state=0)
```

#### <font color='black'> Feature Scaling </font>

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
Y = le.fit_transform(Y)
```

## <font color ='black'> Fitting a Decision-Tree algorithm to the Training set </font>

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier = DecisionTreeClassifier(criterion='entropy', random_state=0)
```

```python
classifier.fit(X,Y)
```

```python
from sklearn.tree import plot_tree

mtp.figure(figsize=(20,20))
tree_img= plot_tree(classifier, feature_names=df.columns, class_names=df['Species'].unique().tolist(), precision=4,label="all",filled=True)
mtp.show()
```
